```python
import docx

# Open the Word document
doc = docx.Document(r"C:\Users\harsh\Jupiter\Resume (2) (2).docx")

# Create a list to store the extracted text
text = []

# Extract all the paragraphs from the document
for para in doc.paragraphs:
    # Split the paragraph text by newline character
    lines = para.text.split('\n')
    # Remove any empty lines and extra spaces
    lines = [line.strip() for line in lines if line.strip()]
    # Add the lines to the list
    text.extend(lines)

# Join the lines into a single string
clean_text = '\n'.join(text)

# Print the extracted and cleaned text
print(clean_text)
```

```python
def extract_info(text):
    # Define regular expressions to match sensitive information
    
    # Pattern for the Dates 
    date_formats = (r"\d{2}-\d{2}-\d{4}", r"\d{2}/\d{2}/\d{4}", r"\d{2}/\d{1}/\d{4}", r"\d{2}[.]\d{2}[.]\d{4}", r"\d{2}-\w{3}-\d{4}", r"\d{2}/\w{3}/\d{4}")
    
    # Pattern for the E-mail
    email_pattern = r"[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-z]{2,}"

    # Find matches and replace with placeholders
    redacted_text = text
    
    # Redacting the Dates
    for date_format in date_formats:
        redacted_text = re.sub(date_format, "XXXXXXXX", redacted_text)
    # Redacting the contact numbers    
    redacted_text = re.sub(r"\d{10}", "XXXXXXX", redacted_text)
    
    # Redacting the E-mail
    redacted_text = re.sub(email_pattern, "XXXXXXXX", redacted_text)

    # Print redacted output
    print(redacted_text)


    
word = extract_info(text)
print(word)
```
