### Task : #1 Cleaning the DATA

#### Performing the mean() for replacing the Zero(0) values in the SkinThickness Column

```python
y = df["SkinThickness"].mean() # Using the mean() 

x = round(y,1)  # Range for the float value

# Replacing the 0 values
df["SkinThickness"] = df["SkinThickness"].replace(0,x)

# The first 5 rows will display
df.head()
```

####  Performing the median() on Blood Pressure Column to replace the 0 values

```python
# Here we use median() because we are getting the normal value of the blood presure ('72')

y = df["BloodPressure"]. median()

x = round(y,0)  # Range for the float value

# Replacing the zero values 
df["BloodPressure"] = df["BloodPressure"].replace(0,x)

# Display the first 10 rows in the table
df.head(10)
# Here if we observe the value of the 7th row in blood pressure is changed
```

#### Performing the mean on zero(0) values in Glucose Column

```python
y = df['Glucose'].mean() # Using mean() to replace the zeros in the glucose

x = round(y,0) # The other way to change the float values into int 

df['Glucose'] = df['Glucose'].replace(0,x) # Replacing the zero values 
                                           # Here we can't use the fillna() because it's mainly used to replace the nullvalues not zeros

df.head() # Displays the first 5 rows
```

### Task : #2 Replacing the Outcome Column values on the base of Glucose Column

##### :Zero(0) represents  No Diabetic
##### :One(1) represents  Diabetic
##### :Two(2) represents  Prediabetic


```python
df['Glucose'].dtype # To check whether the datatype is in int or not 
```

```python
# According to the google the person having Blood sugar level('Glucose') is more than 125 they are Diabetic , persons having Blood sugar level('Glucose') less than 99 they are normal
# and the person who is in the range between 99 and 125 they are considerd as prediabetic

df.loc[ df['Glucose'] >= 125,'Outcome'] = 1 # The Outcome value is changed to 1 if the glucose value is '>'(greater than) 125 

df.loc[ df['Glucose'] < 99,'Outcome'] = 0 # The Outcome value is changed to 0 if the glucose value is '<'(less than) 99

df.loc[(df['Glucose'] >= 99) & (df['Glucose'] < 125),'Outcome'] = 2 # The Outcome value is changed to 2 if the glucose value is '<'(less than) 125 and '>'(greater than) 99
# Note: In pandas we can't use the 'and' operater for comparing (we won't get the proper output) instead use the '&' Symbol 

df.head(10) # Displays the first 10 rows
```

### Task : #3 Changing the Insulin values 

```python
df['Insulin'].dtype # Checking the DataType of the Insulin, if it is in the object then convert it into int
```

```python
# Finding the mean for the Insulin column

# NOte: Here the Outcome is one which means Diabetic So,there is no use of Insulin for Normal and Prediabetic

df.loc[df['Outcome'] == 1, 'Insulin'] =  df['Insulin'].mean() # The Diabetic values are replaced with mean()

df.loc[df['Outcome'] == 0,'Insulin'] = 0 # Normal persons

df.loc[df['Outcome'] == 2,'Insulin'] = 0 # PreDiabetic Persons


df['Insulin'] = df['Insulin'].astype('int64') # Converting the float values into the integers
df.head(20)
# Displays the first 20 rows 
```
