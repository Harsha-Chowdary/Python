<p>
There are many possibilities of kidney to get efficeted by stones. Lets discuss about some major possibilities .

<b>Possibilities</b>
    <li> Not taking <b>water</b> properly in our day to day life.</li>
    <li>Consuming the food which contain heavy <b>salt and sugar</b> in it.</li>
    <li>Not letting the waste of the body out through <b>Urine</b>.</li>
    
<b>Explanation of the each column:</b>
  <ol>
      
  <li><b>Gravity:</b>The specific gravity of urine can provide information about the concentration of minerals and other substances in the urine, which can contribute to the formation of kidney stones. High specific gravity may indicate a risk of stone formation. The <b>Urine Specific Gravity (USG)</b> need to be in the range of <b><u>1.008-1.020</u></b>, if it is higher then <b>1.020</b> then there is a chance of getting stones in the kidney.</li><br>

  <li><b>pH level:</b>The pH level of urine can also provide information about the likelihood of stone formation. If urine is too acidic or alkaline, it can lead to the formation of certain types of stones. SO here the pH level need to be <b>Neutral</b> according to the pH scale neutral means between <b><u>6.6-7.5</u></b>.</li><br>
      
      
  <li><b>Osmolality:</b> Osmolality is a measure of the concentration of particles in urine. High osmolality may indicate that the urine is concentrated and may contribute to stone formation. The Osmolality range of human kidney is <u><b>275 to 295 mOsm/kg</b></u> for adults and older adults.</li> <br>
      
      
  <li><b>Conductivity:</b> Conductivity is a measure of the ability of urine to conduct electricity. High conductivity may indicate the presence of minerals in the urine that can contribute to the formation of stones.</li><br>
      
      
   <li><b>Urea:</b> Urea is a waste product that is excreted in urine. High levels of urea in urine may indicate dehydration, which can lead to concentrated urine and increase the risk of stone formation.</li><br>
      
      
  <li><b>Calcium:</b> Calcium is a mineral that can contribute to the formation of kidney stones. High levels of calcium in urine may indicate an increased risk of stone formation.</li><br>
      
  <li><b>Target:</b> Here in target, <b>Zero(0)</b> means the patient is not containing the stones in kidney and <b>One(1)</b> means the person is contaning stones in the kidney.</li><br>
    
</ol>
</p>

### Changing the Gravity column values on the base of Target column values

<li>If the target value is zero (0) in the dataset, it indicates that the person is normal. For a normal person, the gravity value should fall between <b>1.008 and 1.020</b>.

```python
target = kidney_df['target']

gravity = kidney_df['gravity']

for i in range(len(target)):
    if target[i] == 0 and (gravity[i] < 1.008 or gravity[i] > 1.020):
        gravity[i] = round(random.uniform(1.008, 1.020), 3)
        
kidney_df.head()
```

Here the values are changed. You can see it in the third row the value of gravity is changed . 

### Changing the pH column values on the base of target column and gravity

<li>If the gravity value is between 1.008 and 1.020 and the target is Zero(0) then the pH value will change in between 6.6 to 7.5</li>
<li> If one of them are not satisfying then remain the pH value same</li>

```python
pH = kidney_df['ph']
for i in range(len(target)):
    if target[i] == 0 and (1.008 <= gravity[i] <= 1.020):
        if not (6.6 <= pH[i] <= 7.5):
            pH[i] = round(random.uniform(6.6, 7.5), 2)
            
kidney_df.head()                        
```

### Changing the Osmolality values on the base of pH column

<li>If the pH value is in between 6.6 to 7.5 then the osmoality valu should change in between 275 to 295

```python
osmo = kidney_df['osmo']
for i in range(len(pH)):
    if 6.6 <= pH[i] <= 7.5:
        if not (275 <= osmo[i] <= 295):
            osmo[i] = random.randint(275, 295)
            
kidney_df.head()            
```

### Finding the best columns for the FineTuning by using the feature selection 

```python
from sklearn.feature_selection import SelectKBest, f_classif
X = kidney_df.drop(['target'], axis=1)
y = kidney_df['target']

# perform feature selection using SelectKBest with f_classif score function
selector = SelectKBest(f_classif, k=4)  # select top 4 features
X_new = selector.fit_transform(X, y)

# get the indices of the selected features
indices = selector.get_support(indices=True)

# get the names of the selected features
selected_features = X.columns[indices].tolist()

print("Selected features:", selected_features)
```

### Model :Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Select the columns for fine-tuning
X = kidney_df[['gravity','ph','osmo','cond','urea','calc']]
y = kidney_df['target']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create logistic regression object
logreg = LogisticRegression()
```

```python
# Accuracy on the Training Data 
# Fit the model to the training data

logreg.fit(X_train, y_train)
# predict the X_train
X_train_predict = logreg.predict(X_train)

training_data_accuracy = accuracy_score(X_train_predict, y_train)
#  Make predictions on the test set
y_pred = logreg.predict(X_test)
```

```python
print("The accuracy score of the training data is: {:.1f}%".format(training_data_accuracy * 100))
```

```python
# Accuracy score on the testing data
# predict the X_test
X_test_predict = logreg.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_predict, y_test)
```

```python
print("The accuracy score of the testing data is: {:.1f}%".format(testing_data_accuracy * 100))
```

 <li>Hence the training and testing accuracy is in almost similar Therefore, it is likely that the our model is well-fit to the data

```python
# create a dataframe that will show our predicted results with the actual ones.
Logistic_values=pd.DataFrame({'Actual_y':y_test,'Predict_Y':y_pred})
Logistic_values.head()
```

```python
# Plot the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
# plot the confusion_matrix
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

### Performing the Cost Function or Mean Squared Error(MSE) for the best accuracy in logistic regression

```python
from sklearn.metrics import mean_squared_error

# Compute mean squared error on the testing data
mse = mean_squared_error(y_test, y_pred)

print("Mean squared error:", "{:.3f}".format(mse))
```

### Random Forest
