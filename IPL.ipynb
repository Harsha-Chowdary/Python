### Task #1: Work on the IPL dataset and Calculate the total runs and output in pictorial representation

### IMPORTING PACKAGES

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
test
```

### Droping the unwanted Columns

```python
cleaned_data = test.drop(['non_boundary','is_wicket','dismissal_kind','player_dismissed','fielder','extras_type'],axis = 1 )

cleaned_data
```

```python
cleaned_data[['batsman_runs', 'extra_runs', 'total_runs']].sum()
```

#### Grouping the runs according to the Team names

```python
# Group the dataset by team and sum the relevant columns
team_runs = cleaned_data.groupby('batting_team')[['batsman_runs', 'extra_runs', 'total_runs']].sum()

# Print the resulting dataframe
print(team_runs)
```

#### Ploting the number of runs played by the team , Showing in the graphical representation

```python
# plot the bar graph
X = team_runs.plot(kind='bar', figsize=(10,8), width=0.8)

font = {'family':'Georgia','color':'teal','size':'22','fontweight':'bold'}

# add title and axis labels
X.set_title('Total Runs Scored by IPL Teams', fontsize=20, fontweight='bold')
X.set_xlabel('Team',fontdict = font)
X.set_ylabel('Runs',fontdict = font)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=65, ha='right')

# show the plot
plt.show()
```

### Showing the same result in the Pie Chat

```python
# Group the dataset by team and sum the relevant columns
team_runs = cleaned_data.groupby('batting_team')[['batsman_runs', 'extra_runs', 'total_runs']].sum()

# Plot the pie chart
plt.figure(figsize=(10,8))
plt.pie(team_runs['total_runs'], labels=team_runs.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 14}, pctdistance=0.8)

# Add title
plt.title('Total Runs Scored by IPL Teams', fontsize=20, fontweight='bold')

# Show the plot
plt.show()
```
